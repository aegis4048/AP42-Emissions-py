```python
def calc_Vv(D, Hvo):
    """
    Vv = Tank vapor space volume (ft^3, Eq 1-3)
    D = Tank diameter (ft)
    Hvo = vapor space outage (ft, Eq 1-16)
    """
    return (np.pi / 4) * D ** 2 * Hvo
```

```python
def calc_Hvo(Hs, Hl, Hro):
    """
    Hs = Tank shell height (ft, given)
    Hl  = Tank liquid height (ft, given, assumed to be half-filled mostly)
    Hro = roof outage (ft, Eq 1-17)
    """
    return Hs - Hl + Hro
```

```python
def calc_F_to_R(F):
    return F + 459.67

def calc_R_to_F(R):
    return R - 459.67

def calc_gal_to_bbl(gal):
    return gal / 42

def calc_bbl_to_gal(bbl):
    return bbl * 42
```

```python

def calc_T_LA(Hs, D, Ta, Tb, alpa_r, alpha_s, I):
    _1st = (0.5 - 0.8 / (4.4 * Hs/D + 3.8)) * Ta
    _2nd = (0.5 + 0.8 / (4.4 * Hs/D + 3.8)) * Tb
    _3rd = (0.021 * alpha_r * I + 0.013 * Hs/D * alpha_s * I) / (4.4 * Hs/D + 3.8)
    return _1st + _2nd + _3rd 
```

```python
def find_timeframe(user_input):
    timeframe = [
        ['Annual', 0],
        ['Jan', 'January', 1],
        ['Feb', 'February', 2],
        ['Mar', 'March', 3],
        ['Apr', 'April', 4],
        ['May', 'May', 5],
        ['Jun', 'June', 6],
        ['Jul', 'July', 7],
        ['Aug', 'August', 8],
        ['Sep', 'September', 9],
        ['Oct', 'October', 10],
        ['Nov', 'November', 11],
        ['Dec', 'December', 12]
    ]

    # Normalize the user input
    if isinstance(user_input, str):
        normalized_input = user_input.capitalize()
    elif isinstance(user_input, int):
        if user_input > 12:
            raise ValueError("Invalid input: The number exceeds 12. There are only 12 months in a year plus 0 for 'Annual'.")
        normalized_input = user_input
    else:
        raise ValueError("Invalid input type")

    # Search for the corresponding timeframe
    for period in timeframe:
        if normalized_input in period:
            return period[0]  # Return the true timeframe
            
    if isinstance(user_input, str):
        raise ValueError("Timeframe not found, check for a typo. Examples of accepted string inputs: 'Sep', 'September', 'Feb', 'February', 'Dec', 'December'.... Alternatively, pass integer inputs [0, 12]. Ex: 0='Annual', 1='Jan', 12='Dec'.")
    
    raise ValueError("Timeframe not found.")

# Example usage
print(find_timeframe(0))  # Output: 'Feb'
#print(find_timeframe(13))          # Raises ValueError
```

```python
file = 'Table 7-1-7 Meteorological data for selected US locations.xlsx'
df = pd.read_excel(file, sheet_name='Data')

# Dictionary of state abbreviations to full names, all in uppercase
state_dict = {
    'AL': 'ALABAMA', 'AK': 'ALASKA', 'AZ': 'ARIZONA', 'AR': 'ARKANSAS', 'CA': 'CALIFORNIA',
    'CO': 'COLORADO', 'CT': 'CONNECTICUT', 'DE': 'DELAWARE', 'FL': 'FLORIDA', 'GA': 'GEORGIA',
    'HI': 'HAWAII', 'ID': 'IDAHO', 'IL': 'ILLINOIS', 'IN': 'INDIANA', 'IA': 'IOWA',
    'KS': 'KANSAS', 'KY': 'KENTUCKY', 'LA': 'LOUISIANA', 'ME': 'MAINE', 'MD': 'MARYLAND',
    'MA': 'MASSACHUSETTS', 'MI': 'MICHIGAN', 'MN': 'MINNESOTA', 'MS': 'MISSISSIPPI', 'MO': 'MISSOURI',
    'MT': 'MONTANA', 'NE': 'NEBRASKA', 'NV': 'NEVADA', 'NH': 'NEW HAMPSHIRE', 'NJ': 'NEW JERSEY',
    'NM': 'NEW MEXICO', 'NY': 'NEW YORK', 'NC': 'NORTH CAROLINA', 'ND': 'NORTH DAKOTA', 'OH': 'OHIO',
    'OK': 'OKLAHOMA', 'OR': 'OREGON', 'PA': 'PENNSYLVANIA', 'RI': 'RHODE ISLAND', 'SC': 'SOUTH CAROLINA',
    'SD': 'SOUTH DAKOTA', 'TN': 'TENNESSEE', 'TX': 'TEXAS', 'UT': 'UTAH', 'VT': 'VERMONT',
    'VA': 'VIRGINIA', 'WA': 'WASHINGTON', 'WV': 'WEST VIRGINIA', 'WI': 'WISCONSIN', 'WY': 'WYOMING'
}


input_location = 'Corpus Christi, TX'  # Example input

# Normalize the input location to uppercase
input_normalized = input_location.upper()

# Initialize variables to store matched state info
matched_state_abbr = None
matched_state_name = None

# Check if the normalized input is a state abbreviation
if input_normalized in state_dict:
    matched_state_abbr = input_normalized
    matched_state_name = state_dict[input_normalized]

# Generate combinations of input parts to match multi-word state names
if not matched_state_name:
    input_parts = input_normalized.split()
    for i in range(1, len(input_parts) + 1):
        for combo in combinations(input_parts, i):
            test_str = ' '.join(combo)
            if test_str in state_dict.values():
                matched_state_name = test_str
                matched_state_abbr = [abbr for abbr, name in state_dict.items() if name == test_str][0]
                break
        if matched_state_name:
            break

# Check if input is an exact match for any location
if input_normalized in df['Location'].str.upper().tolist():
    input_location = ' '.join([s.capitalize() for s in input_location.split(' ')[:-1]] + [input_location.split(' ')[-1].upper()])
    
    location_GEODATA = df[df['Location'] == input_location]
    return location_geodata
    print(f"Exact match found for location: '{input_location}'")

    
    
elif matched_state_abbr:
    locations_in_state = df[df['State'].str.upper() == matched_state_abbr]['Location'].unique()
    if locations_in_state.size > 0:
        locations_list_str = ', '.join(f"'{location}'" for location in locations_in_state)
        raise TypeError(f"State name/abbreviation '{input_location}' -> '{matched_state_abbr}' detected. Full state name is '{matched_state_name}'. List of accepted location parameters for '{matched_state_name}': {locations_list_str}")
    else:
        raise TypeError(f"No data available for the state '{matched_state_name}'.")
else:
    # Fuzzy matching to suggest the closest match
    locations = set(df['Location'].str.upper())
    closest_match, score = process.extractOne(input_normalized, locations)

    if score > 90:
        # Split the closest match into city and state, and format them correctly
        suggested_city, suggested_state = closest_match.rsplit(',', 1)
        suggested_city = suggested_city.strip().title()  # City in title case
        suggested_state = suggested_state.strip().upper()  # State in uppercase
        suggested_closest_match = f"{suggested_city}, {suggested_state}"
        raise TypeError(f"Invalid location parameter: '{input_location}'. Did you mean '{suggested_closest_match}'?")
    else:
        raise TypeError(f"Invalid location parameter: '{input_location}'. For the list of available locations, visit www.something.com")
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.distance import geodesic

# Example DataFrame
data = {
    'State': ['AL', 'AL', 'AL', 'AL', 'AR', 'WY', 'WY', 'WY', 'WY', 'WY'],
    'City': ['Birmingham', 'Huntsville', 'Mobile', 'Montgomery', 'Fort Smith', 'Casper', 'Cheyenne', 'Lander', 'Rock Springs', 'Sheridan'],
    'Location': ['Birmingham, AL', 'Huntsville, AL', 'Mobile, AL', 'Montgomery, AL', 'Fort Smith, AR', 'Casper, WY', 'Cheyenne, WY', 'Lander, WY', 'Rock Springs, WY', 'Sheridan, WY']
}
df = pd.DataFrame(data)

def find_closest_location(input_location):
    geolocator = Nominatim(user_agent="geoapiExercises")

    # Geocode user input
    location = geolocator.geocode(input_location)
    user_coords = (location.latitude, location.longitude)

    # Initialize minimum distance and location
    min_distance = None
    closest_location = None

    # Iterate over the DataFrame locations
    for loc in df['Location']:
        loc_coords = geolocator.geocode(loc)
        distance = geodesic(user_coords, (loc_coords.latitude, loc_coords.longitude)).miles

        if min_distance is None or distance < min_distance:
            min_distance = distance
            closest_location = loc

    return closest_location, min_distance

# Example Usage
user_input = "Houston"
closest_location, distance = find_closest_location(user_input)
print(f"The closest location to {user_input} is {closest_location} at a distance of {distance} miles.")
```
