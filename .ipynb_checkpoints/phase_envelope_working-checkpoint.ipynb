```python
def cas_to_name(cas):
    if not isinstance(cas, str):
        raise ValueError("Input must be a string")    
    return ChemicalConstantsPackage.constants_from_IDs([cas]).names[0]

def name_to_cas(name):
    if not isinstance(name, str):
        raise ValueError("Input must be a string")    
    return ChemicalConstantsPackage.constants_from_IDs([name]).CASs[0]

def K_to_R(K):
    return (K - 273.15) * 1.8 + 491.67

def R_to_K(R):
    return (R - 491.67) / 1.8 + 273.15

def F_to_R(F):
    return F + 459.67

def R_to_F(R):
    return R - 459.67

def C_to_K(C):
    return C + 273.15

def F_to_C(F):
    return (F - 32) * 5 / 9

def K_to_C(K):
    return K - 273.15

def gal_to_bbl(gal):
    return gal / 42

def bbl_to_gal(bbl):
    return bbl * 42

def psi_to_Pa(psi):
    return psi * 6894.745

def bar_to_psi(bar):
    return bar * 14.5038

def Pa_to_bar(Pa):
    return Pa / 100000

def Pa_to_psi(pa):
    return pa / 6894.745

def Pa_to_mmHg(Pa):
    return Pa / 133.322

def mmHg_to_Pa(mmHg):
    return mmHg / 0.0075

def F_to_K(F):
    return (F - 32) * 5/9 + 273.15

def K_to_F(K):
    return (K - 273.15) * 1.8 + 32
```

```python
from thermo import ChemicalConstantsPackage, PRMIX, CEOSLiquid, CEOSGas, PropertyCorrelationsPackage
from thermo import FlashVLN
import matplotlib.pyplot as plt
from thermo.interaction_parameters import IPDB

def psi_to_Pa(psi):
    return psi * 6894.745

def F_to_K(F):
    return (F - 32) * 5/9 + 273.15

def Pa_to_psi(pa):
    return pa / 6894.745

def K_to_F(K):
    return (K - 273.15) * 1.8 + 32

comp = dict([
('methane', 0.2),
('propane', 0.2),
('hexane', .6),
('water', 0.),
])

total_comp = sum(comp.values())
if total_comp > 1:
    comp = {k: v / total_comp for k, v in comp.items()}
    
constants, properties = ChemicalConstantsPackage.from_IDs(comp.keys())
kijs = IPDB.get_ip_asymmetric_matrix('ChemSep PR', constants.CASs, 'kij')

eos_kwargs = dict(Tcs=constants.Tcs, Pcs=constants.Pcs, omegas=constants.omegas, kijs=kijs)
gas = CEOSGas(PRMIX, eos_kwargs, HeatCapacityGases=properties.HeatCapacityGases)
liq = CEOSLiquid(PRMIX, eos_kwargs, HeatCapacityGases=properties.HeatCapacityGases)
flashN = FlashVLN(constants, properties, liquids=[liq, liq], gas=gas) # if dealing with water, fill [liq, liq] twice.

P = psi_to_Pa(100)
T = F_to_K(100)

zs = list(comp.values())

res = flashN.flash(T=T, P=P, zs=zs)

Ts, P_dews, P_bubbles, branch_Ps = flashN.plot_TP(zs=zs, show=False, Tmin=100, Tmax=475, pts=50, hot=True, values=True)
Ts2, P_dews2, P_bubbles2, branch_Ps2 = flashN.plot_TP(zs=zs, show=False, Tmin=260, Tmax=475, pts=50, hot=True, values=True)
```

# C6+, C7+ critical point estimation

```python
from scipy.optimize import newton
import numpy as np

rho = 0.8153  # specific gravity to liquid-phase water
M = 187
omega = 0.5980960919818

mw = 187.8
sg = 0.6456428956368

# notes: SRK is not validated yet. Got wrong results. For now, use only Peng-Robinson

def get_eos_constants(eos):
    
    constants_srk = {
        "c1": 1.6312e2, "c2": 8.6052e1, "c3": 4.3475e-1, "c4": -1.8774e3, 
        "d1": -1.3408e-1, "d2": 2.5019, "d3": 2.0846e2, "d4": -3.9872e3, "d5": 1.0,
        "e1": 7.4310e-1, "e2": 4.8122e-3, "e3": 9.6707e-3, "e4": -3.7184e-6,
        "w1": 0.480, "w2": 1.574, "w3": 0.17, 
    }
    constants_pr = {
        "c1": 7.34043e1, "c2": 9.73562e1, "c3": 6.18744e-1, "c4": -2.05932e3,
        "d1": 7.28462e-2, "d2": 2.18811, "d3": 1.63910e2, "d4": -4.04323e3, "d5": 0.25,
        "e1": 3.73765e-1, "e2": 5.49269e-3, "e3": 1.17934e-2, "e4": -4.93049e-6,  
        "w1": 0.37464, "w2": 1.54226, "w3": 0.26992,         
    }    
    if eos == 'PR':
        return constants_pr
    elif eos == 'SRK':
        return constants_srk
    else:
        raise ValueError("Unsupported EOS. Pick either PR or SRK")   
    
def calc_m(M=None, rho=None, w=None, eos='PR', mode='acentric'):
    c = get_eos_constants(eos)
    if mode == 'mw_density':
        return c['e1'] + c['e2']*M + c['e3']*rho + c['e4']*(M**2)
    elif mode == 'acentric':
        return c['w1'] + c['w2']*w - c['w3']*(w**2)
    else:
        raise ValueError("Unsupported mode. Pick either acentric or mw_density")  

        
# Use of Critical Point in Equation of State Modeling for Reservoir Fluids
# eq 5, 6, and 7
# valid only for carbon-compounds, C1 ~ C80
def calc_m_acentric_obj_func1(w, M, rho, eos='PR'):
    c = get_eos_constants(eos)
    return -calc_m(M=M, rho=rho, mode='mw_density') + c['w1'] + c['w2']*w - c['w3']*(w**2)

# modified accentric factor correlation when w >= 0.5
# file:///C:/Users/EricKim/Documents/AP42/AP42-Emissions-py/spe-174558-pa.pdf
def calc_m_acentric_obj_func2(w, M, rho, eos='PR'):
    c = get_eos_constants(eos)
    return -calc_m(M=M, rho=rho, mode='mw_density') + c['w1'] + c['w2']*w - c['w3']*(w**2) + 0.016666*(w**3)

def calc_acentric_from_mw_sg(mw, sg, eos='PR'):
    omega = newton(calc_m_acentric_obj_func1, x0=0.5, args=(mw, sg, eos))
    if omega >= 0.5 and eos == 'PR':
        print('modified PR')
        omega = newton(calc_m_acentric_obj_func2, x0=0.5, args=(mw, sg, eos))
    return omega

def calc_acentric(Tb=None, Tc=None, Pc=None, mode='edmister'):
    
    # T in rankine, P in psia
    if mode == 'edmister':
        return 3/7 * (np.log10(Pc/14.7)/((Tc/Tb) - 1)) - 1
    
    # this is what promax uses. 
    elif mode == 'lee-kesler':
        
        def lee_kesler(w, Tb, Tc, Pc):
            Tr = Tb / Tc
            Pr = 14.6959 / Pc  # pressure at normal boiling T is assumed to be 1 atm. 
            f0 = 5.92714 - 6.09648 / Tr - 1.28862 * np.log(Tr) + 0.169347 * Tr**6
            f1 = 15.2518 - 15.6875 / Tr - 13.4721 * np.log(Tr) + 0.43577 * Tr**6
            return f0 + w * f1 - np.log(Pr)

        return newton(lee_kesler, x0=0.5, args=(Tb, Tc, Pc))
```

### Pure Pseudo-compound (not mixture) Pc and Tc  -----------------------------------------------------------------------

```python
# Pederson (1992)
def calc_pseudo_Pc(M, rho, eos='PR'):
    # bar
    c = get_eos_constants(eos) 
    return np.exp(c['d1'] + c['d2']*rho**(c['d5']) + c['d3']/M + c['d4']/(M**2))
        
def calc_pseudo_Tc(M, rho, eos='PR'):
    # K
    c = get_eos_constants(eos)   
    return c['c1']*rho + c['c2']*np.log(M) + c['c3']*M + c['c4']/M

# Use of Critical Point in Equation of State Modeling for Reservoir Fluids (2023)
def calc_pseudo_Tc_2(M):
    # SRK
    # Tc in Kalvin
    b1 = 224.92
    b2 = -491.08
    return b1 * np.log(M) + b2

def calc_pseudo_Pc_2(M):
    # SRK
    # Pc in bar
    b3 = 11.3
    b4 = -24.454
    return b3 * np.log(M) + b4

# API data book, Riazi (1985) -> this seems to best-match promax results
def calc_pseudo_Tc_(Tb, sg):
    "Procedure 4D3.1"
    "Tb = boiling T in K"
    "sg = specific gravity (unitless)"
    "return = Tc in K"
    Tb = K_to_R(Tb)
    Tc = 10.6443 * np.exp(-5.1747e-4 * Tb - 0.54444 * sg + 3.5995e-4 * Tb * sg) * Tb**0.81067 * sg**0.53091
    return R_to_K(Tc)

def calc_pseudo_Pc_(Tb, sg):
    "Procedure 4D4.1"
    "Tb = boiling T in K"
    "sg = specific gravity (unitless)"
    "return = Pc in psia"
    Tb = K_to_R(Tb)
    return 6.161e6 * np.exp(-4.725e-3 * Tb - 4.8014 * sg + 3.1939e-3 * Tb * sg) * Tb**(-0.4844) * sg**4.0846
```

### Vapor Pressure Calculation for Pure Pseudo-component:

https://en.wikipedia.org/wiki/Lee%E2%80%93Kesler_method

This method works for pure pseudo-comps.

```python
def dummy(Tc, Pc, T, w):
    Tr = T / Tc
    f0 = 5.92714 - 6.09648 / Tr - 1.28862 * np.log(Tr) + 0.169347 * Tr**6
    f1 = 15.2518 - 15.6875 / Tr - 13.4721 * np.log(Tr) + 0.43577 * Tr**6
    return np.exp(f0 + w * f1) * Pc

mw = 187.8
sg = 0.6456428956368
Tc = 692.9649506843
Pc = 281.2763251747
T = 588.7
omega = 0.598

dummy(Tc, Pc, T, omega)
```

### Boiling point vs MW Correation --------------------------

```python
mw = 187
sg = 0.8
# Promax Tb = 511.363K

def solve_Tb(mw=None, sg=None, model='riazi_1986'):
    
    # API Databook, 2B2.1
    if model == 'riazi_1986':
        def obj_func(Tb, mw, sg):
            return 20.486 * np.exp(1.165e-4*Tb - 7.78712*sg + 1.1582e-3*Tb*sg) * Tb**1.26007 * sg**4.98308 - mw
        return R_to_K(newton(obj_func, x0=300, args=(mw, sg)))
    
    # Measurement and Modeling of Solubility and Saturated-Liquid Density and Viscosity for Methane/Athabasca-Bitumen Mixtures
    # SPE-174558
    # eq 2, Tb and MW correlation for C7~100, Riazi (2005)
    elif model == 'riazi_2005':
        def obj_func(Tb, mw):
            return ((6.97996 - np.log(1080 - Tb))/0.01964)**(3/2) - mw      
        return newton(obj_func, x0=300, args=(mw, )) # Tb by default is in unit of K

    else:
        raise ValueError('Unsupported mode. Supported modes are: ["riazi_1986", "riazi_2005"]')  
```
